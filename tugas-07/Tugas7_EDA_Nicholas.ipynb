##  High-Impact Insights untuk Pipeline Data

### 1. Data Sangat Skewed → Wajib Transformasi Log

Distribusi kolom seperti `TotalCases`, `Population`, dan `TotalTests` sangat timpang (median vs max berbeda ratusan kali lipat). Model linear dan distance-based seperti KNN atau clustering akan bias ke negara besar jika tidak ditangani.

> `log1p` dipilih karena aman untuk nilai nol dan mengompresi nilai ekstrem.

***

### 2. Kolom `New*` → Drop Langsung

`NewCases`, `NewDeaths`, `NewRecovered` memiliki **>98% missing value** — bukan sekadar data hilang, tapi memang tidak tersedia secara struktural. Dataset ini adalah **snapshot statis**, bukan time-series.

***

### 3. Feature Redundancy → Pilih Satu Representasi

Kolom seperti `Tot Cases/1M pop` adalah hasil bagi matematis dari `TotalCases / Population`. Menyimpan keduanya menyebabkan **multikolinearitas**.

| Option | Kolom yang Dipertahankan | Cocok untuk |
|---|---|---|
| **A** | `TotalCases`, `TotalDeaths`, `TotalTests` | Analisis volume absolut |
| **B** | `Tot Cases/1M pop`, `Deaths/1M pop`, `Tests/1M pop` | Perbandingan antar negara |

Untuk **cross-country comparison**, pilih **Option B** (per 1M pop lebih adil).

***

### 4. Missing Value Tidak Random → Imputasi Tepat Sasaran

Missing bukan karena data tidak diambil, melainkan karena **negara kecil dan konflik tidak melapor**. Mean imputation bisa menyesatkan.


***

### 5. Dataset Kecil (209 baris) → Jaga Kompleksitas Model

Dengan hanya 209 observasi, hindari model yang terlalu kompleks. Pilihan terbaik:

- Linear / Ridge / Lasso Regression
- Decision Tree (kedalaman dangkal)
- K-Means Clustering sederhana

Overengineered:

- Deep Learning
- Ensemble model berat (XGBoost dengan banyak fitur)

***

### 6. Encoding Kolom Kategorikal

`Continent` dan `WHO Region` bisa digunakan untuk segmentasi atau sebagai fitur model.

**5. Death Rate & Recovery Rate**
> Buat kolom baru "Death Rate" = TotalDeaths / TotalCases

> Buat kolom "Recovery Rate" = TotalRecovered / TotalCases

> Bandingkan death rate dan recovery rate secara global

```python
# A. Membuat kolom baru

df["Death Rate"] = df["TotalDeaths"] / df["TotalCases"]
df["Recovery Rate"] = df["TotalRecovered"] / df["TotalCases"]

# cek hasil
df[[
    "Country/Region",
    "TotalCases",
    "TotalDeaths",
    "TotalRecovered",
    "Death Rate",
    "Recovery Rate"
]].head()
```

```python
# Hitung nilai global (rata-rata seluruh negara)

global_death_rate = df["Death Rate"].mean()
global_recovery_rate = df["Recovery Rate"].mean()

print("Global Death Rate:", global_death_rate)
print("Global Recovery Rate:", global_recovery_rate)
```

```python
# Visualisasi perbandingan global

labels = ["Death Rate", "Recovery Rate"]
values = [global_death_rate, global_recovery_rate]

plt.figure()

plt.bar(labels, values)

plt.title("Perbandingan Global Death Rate vs Recovery Rate")
plt.ylabel("Rate")

plt.show()
```

Insight :

1. Recovery rate global jauh lebih tinggi dibanding death rate

Berdasarkan perhitungan:

- Global Death Rate ≈ 3.27%
- Global Recovery Rate ≈ 71.25%

Berdasarkan hasil eksplorasi dataset COVID-19 dari Worldometer yang mencakup 209 negara, ditemukan bahwa penyebaran pandemi tidak merata di seluruh dunia. Sebagian besar negara memiliki jumlah kasus relatif kecil, sementara beberapa negara dengan populasi besar seperti USA, Brazil, dan India memiliki jumlah kasus yang jauh lebih tinggi dibanding negara lain. Hal ini menunjukkan adanya ketimpangan distribusi dampak pandemi antar negara.

Analisis distribusi menunjukkan bahwa data kasus, kematian, dan kesembuhan memiliki pola yang tidak seimbang (skewed), dimana hanya sedikit negara yang memiliki nilai sangat tinggi. Oleh karena itu, penggunaan skala log membantu memperjelas pola distribusi data. Selain itu, terdapat kecenderungan bahwa negara dengan populasi lebih besar memiliki jumlah kasus yang lebih tinggi, meskipun populasi bukan satu-satunya faktor yang mempengaruhi penyebaran COVID-19.

Analisis per benua menunjukkan bahwa North America, Asia, dan South America memiliki jumlah kasus tertinggi, sedangkan Australia/Oceania memiliki jumlah kasus paling rendah. Hal ini menunjukkan bahwa dampak pandemi berbeda-beda antar wilayah geografis.

Dari sisi testing, ditemukan bahwa jumlah tes berkorelasi positif dengan jumlah kasus yang terdeteksi. Negara dengan jumlah testing yang tinggi cenderung memiliki jumlah kasus terkonfirmasi lebih banyak. Positivity rate yang tinggi pada beberapa negara menunjukkan kemungkinan keterbatasan testing atau tingginya tingkat penyebaran virus.

Analisis death rate dan recovery rate menunjukkan bahwa secara global recovery rate (sekitar 71%) jauh lebih tinggi dibanding death rate (sekitar 3%). Hal ini menunjukkan bahwa sebagian besar pasien COVID-19 berhasil sembuh. Meskipun demikian, masih terdapat kasus aktif yang menunjukkan pandemi masih berlangsung pada saat data dikumpulkan.

Analisis kasus serious/critical menunjukkan bahwa negara dengan jumlah kasus aktif tinggi cenderung memiliki jumlah kasus kritis yang lebih banyak. Namun, proporsi kasus kritis tetap relatif kecil dibanding total kasus aktif.

Perbandingan performa negara menunjukkan bahwa beberapa negara memiliki recovery rate sangat tinggi, terutama negara dengan jumlah kasus relatif kecil. Sebaliknya, beberapa negara memiliki death rate yang cukup tinggi dibanding negara lain. Negara dengan testing per 1 juta penduduk tertinggi umumnya merupakan negara dengan populasi kecil, sehingga lebih mudah mencapai angka testing yang tinggi secara proporsional.

Berdasarkan correlation matrix, terdapat hubungan kuat antara jumlah kasus dengan jumlah kematian dan kesembuhan. Selain itu, jumlah tes dan populasi juga berkorelasi dengan jumlah kasus, yang menunjukkan bahwa testing dan ukuran populasi berperan dalam jumlah kasus yang terdeteksi.

Secara keseluruhan, EDA ini menunjukkan bahwa faktor populasi, testing, dan kondisi geografis memiliki hubungan dengan jumlah kasus COVID-19 yang terdeteksi. Dataset ini dapat digunakan untuk memahami pola penyebaran pandemi secara global, meskipun terdapat keterbatasan data seperti missing values dan tidak tersedianya data time-series.
